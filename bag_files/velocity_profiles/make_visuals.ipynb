```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
filefolder = os.getcwd()
print(filefolder)
# print(os.listdir(filefolder))
ext = ('.csv')
```

```python
for files in os.listdir(filefolder):
    if files.endswith(ext):
        try:
            data = pd.read_csv(files)
        except:
            continue
        title = os.path.splitext(files)[0]
        ax=data.plot.scatter(x='%time',y='field.data')
        ax.set_xlabel('Time')
        ax.set_ylabel('Velocity (m/s)')
        ax.set_title(title)
        new_img = filefolder+'/plot_img/'+title+'.png'
        plt.savefig(new_img)

        # plt.show()
```
