```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
filefolder = os.getcwd()
print(filefolder)
# print(os.listdir(filefolder))
ext = ('.csv')
```

```python
for files in os.listdir(filefolder):
    print(files)
    if files.endswith(ext):
        try:
            data = pd.read_csv(files)
        except:
            continue
        data = data[450:850]
        title = os.path.splitext(files)[0]
        ax=data.plot.line(x='%time',y='field.data')
        ax.set_xlabel('Time')
        ax.set_ylabel('Error (rad)')
        ax.set_title('Odometry Angle Error from True Value with Gaussian Noise')
        new_img = filefolder+'/plot_img/'+title+'.png'
        plt.savefig(new_img)

        # plt.show()
```
