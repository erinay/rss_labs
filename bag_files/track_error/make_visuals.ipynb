```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
filefolder = os.getcwd()
print(filefolder)
# print(os.listdir(filefolder))
ext = ('.csv')
```

```python
for files in os.listdir(filefolder):
    if files.endswith(ext):
        data = pd.read_csv(files)
        title = os.path.splitext(files)[0]
        ax=data.plot.scatter(x='%time',y='field.data')
        ax.set_xlabel('Time')
        ax.set_ylabel('Error Distance from Center(m)')
        ax.set_title(title)
        ax.set_title('Full Lap Data')

        img_file_loc= os.path.splitext(os.path.abspath(files))[:-1]
        # plt.savefig('')

        # plt.show()
```
